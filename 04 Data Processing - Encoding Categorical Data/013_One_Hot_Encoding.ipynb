**One-hot encoding** is a technique used to convert categorical data into a format that can be provided to machine learning algorithms to improve performance. This method transforms each category of a feature into a new binary column, where a "1" indicates the presence of a category, and a "0" indicates its absence.

For example, if you have a feature representing colors with three possible values ("red," "blue," "green"), one-hot encoding will create three new binary columns as follows:

Screenshot 2024-10-12 at 1.48.06 AM.png


This method prevents algorithms from assuming a relationship between the categories, such as one being "greater" or "lesser" than another.

# **OneHotEncoding using Sklearn**

```python
from sklearn.model_selection import train_test_split
```

```python
# Select features (X) and target variable (y)
X = df[['age','workclass', 'education', 'occupation']]  # Input features
y = df['income']
```

```python
# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['workclass','occupation']])
```

```python
X_test_new = ohe.transform(X_test[['workclass','occupation']])
```

```python
X_train_new.shape
```

```python
np.hstack((X_train[['education','age']].values,X_train_new))
```

**OneHotEncoding with Top Categories**

```python
counts = df['occupation'].value_counts()
```

```python
df['occupation'].nunique()
threshold = 2000
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['occupation'].replace(repl, 'other')).astype(int).sample(5)
```
