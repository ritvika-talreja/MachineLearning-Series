**The goal of EDA** is to summarize the main features of the dataset and uncover patterns, anomalies, relationships, and trends in the data. This is achieved through statistical graphics, plots, and other visual tools.

**Bivariate analysis:** It is used to examine the relationship between two variables. By exploring how two features interact, we can gain insights into potential correlations or dependencies that can inform further modeling or decision-making.

**Multivariate analysis:** It goes beyond bivariate analysis by examining relationships between more than two variables simultaneously. It allows us to explore more complex interactions and dependencies in the dataset, providing a comprehensive understanding of how multiple features contribute to the outcome.

This notebook performs bivariate analysis on the **Tips dataset** to explore relationships between two variables, such as how total bills and tips vary based on gender, smoking status, and party size. It also conducts multivariate analysis on the **World Happiness Report** to examine how multiple factors, like GDP, social support, and life expectancy, collectively influence happiness scores.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Bivariate Analysis**

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

This scatter plot shows the relationship between the 'total bill' and 'tip'. It uses the 'sex' column to differentiate the points by color, 'smoker' to vary the point style, and 'size' to vary the size of the points. This allows us to examine the relationship in more detail.

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)

plt.title("Bivariate Analysis: Total Bill vs Tip with Hue, Style, and Size")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend()
plt.show()
```

# **Multivariate Analysis**

```python
data = pd.read_csv("world_happiness_report.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
print("\nMissing Values:\n", data.isnull().sum())
```

```python
data.describe()
```

This heatmap represents the correlation between various numerical features in the dataset, showing how strongly features like 'Happiness Score', 'GDP per capita', 'Life Expectancy', etc are related to each other. Positive correlations are shown in warmer colors, while negative correlations are in cooler colors.

```python
# Filter only numerical columns for correlation
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Correlation matrix for numerical features
correlation_matrix = numerical_data.corr()

# Heatmap of correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - World Happiness Report")
plt.show()
```

The pairplot below visualizes the relationships between key variables like 'Ladder score'(Happiness Score), 'GDP per capita', 'Social support', and others. It helps in understanding how these features are related to each other across different countries.

```python
# Pairplot for key features
selected_features = [
    'Ladder score',
    'Logged GDP per capita',
    'Social support',
    'Healthy life expectancy',
    'Freedom to make life choices',
    'Generosity'
]
sns.pairplot(data[selected_features], diag_kind='kde', markers='o', height=2.5)
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

This scatter plot shows the relationship between 'Logged GDP per capita' and 'Ladder score'(Happiness Score). It helps in visualizing how GDP per capita influences happiness, with different regions color-coded to see regional patterns.

```python
# Scatter plot:GDP vs Ladder Score
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Logged GDP per capita',
    y='Ladder score',
    hue='Regional indicator',
    data=data,
    palette='viridis',
    s=100
)
plt.title("Happiness Score vs GDP per Capita")
plt.xlabel("Logged GDP per Capita")
plt.ylabel("Happiness Score (Ladder Score)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

The boxplot below shows the distribution of happiness scores ('Ladder score') across different regions. It helps in identifying how happiness is distributed across various regions, highlighting differences in median scores and variability.

```python
# Boxplot for Ladder Score across Regions
plt.figure(figsize=(14, 8))
sns.boxplot(x='Regional indicator', y='Ladder score', data=data, palette='Set2')
plt.title("Distribution of Happiness Score by Region")
plt.xticks(rotation=45)
plt.show()
```

This bar plot displays the top 10 countries with the highest happiness scores ('Ladder score'). It provides a quick view of which countries are perceived as the happiest based on the World Happiness Report.

```python
# Bar plot: Top 10 Happiest Countries
top_10_happiest = data.nlargest(10, 'Ladder score')
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Country name',
    y='Ladder score',
    data=top_10_happiest,
    palette='Spectral'
)
plt.title("Top 10 Happiest Countries")
plt.ylabel("Happiness Score (Ladder Score)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()
```

This bar plot shows the bottom 10 countries with the lowest happiness scores ('Ladder score'). It helps to identify which countries are the least happy according to the World Happiness Report.

```python
# Bar plot: Bottom 10 Least Happy Countries
bottom_10_least_happy = data.nsmallest(10, 'Ladder score')
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Country name',
    y='Ladder score',
    data=bottom_10_least_happy,
    palette='coolwarm'
)
plt.title("Bottom 10 Least Happy Countries")
plt.ylabel("Happiness Score (Ladder Score)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()
```
